## Czy wielokąt jest $y$-monotoniczny?

Wielokąt jest monotoniczny, gdy jego wierzchołki mogą być ułożone w taki sposób, że jedna z jego współrzędnych (na przykład współrzędna $x$ lub $y$, w zależności od układu współrzędnych) zawsze rośnie lub maleje wzdłuż kolejnych wierzchołków. Innymi słowy, dla każdej pary wierzchołków wielokąta (oprócz wierzchołka startowego i końcowego), jeden z punktów ma większą (lub mniejszą) wartość danej współrzędnej niż drugi punkt.

W praktyce, wielokąt monotoniczny może być łatwiej sortowany lub przetwarzany w pewnych algorytmach geometrycznych, ponieważ istnieje pewna kolejność, w jakiej wierzchołki pojawiają się wzdłuż danej osi (np. osi $x$ lub $y$). Monotoniczność może ułatwić znajdowanie przecięć linii w takim wielokącie lub wykonywanie innych operacji geometrycznych. W tym zadaniu interesuje nas monotoniczność wielokąta wzdłuż osi $y$.

<span style="color:red">Ćw.</span> Uzupełnij funkcję ```is_y_monotonic```. Pamiętaj, aby sprawozdanie zawierało krótki opis działania tej funkcji.

````python
def find_min_max(polygon):
    n=len(polygon)
    y_min = polygon[0][1]
    y_max = polygon[0][1]
    y_min_index = 0
    y_max_index = 0
    for i in range(1,n):
        if polygon[i][1]<y_min:
            y_min_index = i
            y_min=polygon[i][1]
        if polygon[i][1]>y_max:
            y_max_index=i
            y_max=polygon[i][1]
    return y_min_index, y_max_index #standardowe wyszukiwanie maxa i mina z ich indeksami
    
def l_r_l(polygon): #dzielnie na lancuchy lewy i prawy
    n=len(polygon)
    left=[]
    right=[]
    y_min_index, y_max_index = find_min_max(polygon) #szukamy indexow maxa i mina
    if y_min_index < y_max_index:
        #nalezy zauwazyc ze dodajemy do lancuchow krotke (x, y) oraz indeks wystepowania w oryginalnym polygonie 
        #oraz fakt tego ze lancuchy beda posortowane po y od najwiekszych do najmniejszych
        for i in range(y_max_index-1,y_min_index-1,-1):
            right.append((polygon[i],i))
        for i in range(y_max_index,n):
            left.append((polygon[i],i))
        for i in range(y_min_index):
            left.append((polygon[i],i))
    else:
        
        for i in range(y_max_index,y_min_index):
            left.append((polygon[i],i))
        for i in range(y_max_index-1,-1,-1):
            right.append((polygon[i],i))
        for i in range(n-1,y_min_index-1,-1):
            right.append((polygon[i],i))
            
    return left, right
def is_y_monotonic(polygon):
    lewaczesc,prawaczesc = l_r_l(polygon)
    l=len(lewaczesc)
    r=len(prawaczesc)
    for i in range(l-1):
        if lewaczesc[i][0][1]<lewaczesc[i+1][0][1]:
            return False
    for i in range(r-1):
        if prawaczesc[i][0][1]<prawaczesc[i+1][0][1]:
            return False
    return True
````

# Podział wierzchołków na kategorie

Wierzchołki naszego wielokąta możemy podzielić na parę kategorii:
- początkowe, gdy obaj jego sąsiedzi leżą poniżej i kąt wewnętrzny ma mniej niż 180 stopni. To wierzchołki, w których zaczyna się monotoniczny spadek
- końcowe, gdy obaj jego sąsiedzi leżą powyżej i kąt wewnętrzny ma mniej niż 180 stopni. To wierzchołki, w których monotoniczność wielokąta się zmienia, czyli na przykład zaczyna się monotoniczny wzrost, jeśli wcześniej był spadek, lub na odwrót.

Wierzchołki startowe i końcowe są ważne w kontekście algorytmów przetwarzania wielokątów monotonicznych, takich jak algorytmy dziel i zwyciężaj oraz triangulacji.
- dzielący, gdy obaj jego sąsiedzi leżą poniżej i kąt wewnęntrzny ma więcej niż 180 stopni. To wierzchołki, które wyznaczają przekątne (linie łączące), tworzące trójkąty podczas triangulacji.
- łączący, gdy obaj jego sąsiedzi leżą powyżej i kąt wewnęntrzny ma więcej niż 180 stopni. To wierzchołki, które są połączone liniami (przekątnymi) wewnątrz wielokąta, tworząc trójkąty.

Wierzchołki łączące i dzielące odgrywają kluczową rolę w procesie triangulacji wielokątów, pozwalając na podział figury na trójkąty w sposób bezkolizyjny.
- prawidłowy, pozostałe przypadki, jeden sąsiad powyżej, drugi poniżej

<span style="color:red">Ćw.</span> Uzupełnij funkcję ```color_vertex``` .

````python
def angle_n_height(v1, v2, v3):
    ab = (v2[0] - v1[0], v2[1] - v1[1])
    bc = (v3[0] - v2[0], v3[1] - v2[1])

    determinant = ab[0] * bc[1] - ab[1] * bc[0]
 
    if determinant > 0:
        orientation = 0 #pozniej pi
    elif determinant < 0:
        orientation = 1 #powyzej pi
    else:
        orientation = 2 #collinearne
    
    return orientation
def give_number(v1, v2, v3, orientation):
    if orientation == 0 and v1[1] < v2[1] and v3[1] < v2[1]:
        return 0 #poczatkowy
    elif orientation == 0 and  v1[1] > v2[1] and v3[1] > v2[1]:
        return 1 #koncowy
        
    elif orientation == 1 and  v1[1] > v2[1] and v3[1] > v2[1]:
        return 2 #laczacy
        
    elif  orientation == 1 and  v1[1] < v2[1] and v3[1] < v2[1]:
        return 3 #dzielacy
    else:
        return 4#prawidlowy

    
````

````python
def color_vertex(polygon):
    """
    Funkcja dzieli wierzchołki na kategorie i przypisuje wierzchołkom odpowiednie numery: 0 - początkowy, 1 - końcowy, 2 - łączący, 3 - dzielący, 4 - prawdiłowy
    :param polygon: tablica krotek punktów na płaszczyźnie euklidesowej podanych przeciwnie do ruchu wskazówek zegara - nasz wielokąt
    :return: tablica o długości n, gdzie n = len(polygon), zawierająca cyfry z przedziału 0 - 4, gdzie T[i] odpowiada kategorii i-tego wierzchołka.
    """
    n = len(polygon)
    numer = [0 for i in range(n)]
    numer[n-1] = give_number(polygon[n-2], polygon[n-1], polygon[0], angle_n_height(polygon[n-2], polygon[n-1], polygon[0]))
    numer[0] = give_number(polygon[n-1], polygon[0], polygon[1], angle_n_height(polygon[n-1], polygon[0], polygon[1]))
    for i in range(1, n-1):
        orientation = angle_n_height(polygon[i-1], polygon[i], polygon[i+1])
        numer[i] = give_number(polygon[i-1], polygon[i], polygon[i+1], orientation)
    return numer
    
````

Triangulacja wielokąta monotonicznego to proces podziału wielokąta monotonicznego na trójkąty poprzez dodawanie przekątnych (linii łączących wierzchołki), które nie przecinają się wewnętrznie.

<span style="color:red">Ćw.</span> Uzupełnij funkcję ```triangulation```. Wykorzystaj algorytm opisany na wykładzie.

````python
def l_r_l(polygon):#funkcja dzieli na lewy oraz prawy lancuch
    n=len(polygon)
    left=[]
    right=[]
    y_min_index, y_max_index = find_min_max(polygon) #indexy maks i min wierzcholka
    if y_min_index < y_max_index: #dla sytuacji w ktorej wczesniej jest w polygonie index mniejszego y-ka od wiekszego
        
        for i in range(y_max_index-1,y_min_index-1,-1): #przejscie od ymax_index do y-min_index i dodanie krotki (x, y) oraz indeksu z polygona 
            right.append((polygon[i],i))
        for i in range(y_max_index,n): #od y-maxindex do konca i dodajemy do lewego lancucha (x, y) i index z polygona
            left.append((polygon[i],i))
        for i in range(y_min_index):#od 0 do min_y_index i dodajemy do lewego lancucha (x, y) i index z polygona
            left.append((polygon[i],i))
    else:
        
        for i in range(y_max_index,y_min_index):
            left.append((polygon[i],i))
        for i in range(y_max_index-1,-1,-1):
            right.append((polygon[i],i))
        for i in range(n-1,y_min_index-1,-1):
            right.append((polygon[i],i))
            
    return left, right #zwracamy 2 lancuchy ktore sa juz posortowane od y-max do y-min w postaci listy krotek [((x, y), index w polygon), ...]

def porzadkuj(lewaczesc, prawaczesc, n):
    dl_lewej = len(lewaczesc)
    dl_prawej = len(prawaczesc)
    uporzadkowana_lista = [None for _ in range(n)]
    
    indeks_lewy = indeks_prawy = 0
    while indeks_lewy < dl_lewej and indeks_prawy < dl_prawej:
        if lewaczesc[indeks_lewy][0][1] > prawaczesc[indeks_prawy][0][1]: #jezeli y z lewej czesci jest wiekszy od y z prawej czesci
            uporzadkowana_lista[indeks_lewy + indeks_prawy] = (
                lewaczesc[indeks_lewy][0],
                lewaczesc[indeks_lewy][1],
                'l'
            ) #wstawiamy krotke (x, y, "z  ktorego lancucha dodane")
            indeks_lewy += 1
        else:
            uporzadkowana_lista[indeks_lewy + indeks_prawy] = (
                prawaczesc[indeks_prawy][0],
                prawaczesc[indeks_prawy][1],'r')
            indeks_prawy += 1
    while indeks_lewy < dl_lewej:
        uporzadkowana_lista[indeks_lewy + indeks_prawy] = (lewaczesc[indeks_lewy][0],lewaczesc[indeks_lewy][1],'l')
        indeks_lewy += 1
    while indeks_prawy < dl_prawej:
        uporzadkowana_lista[indeks_lewy + indeks_prawy] = (prawaczesc[indeks_prawy][0],prawaczesc[indeks_prawy][1],'r')
        indeks_prawy += 1
    return uporzadkowana_lista #zwracamy uporzadkowana liste krotek [((x, y), index w polygon,"ktory to lancuch"), ...]
    

def trojkatdobry(wierz1, wierz2, aktualny_wierz, rozmiar): #sprawdzamy czy dany trojkat z danych punktow nalezy do wielokata
    if abs(aktualny_wierz[1] - wierz1[1]) in (1, rozmiar - 1):#gdy mamy sytuacje ze oba wierzcholki sa saiadami tj roznica indeksow == 1 lub n- 1
        return False #zwracamy wtedy falsz bo taki trojkat nie nalezy do wielokata
    if aktualny_wierz[2] == 'r': #jesli aktuany wierzcholek jest z prawego lanucha 
        return angle_n_height(aktualny_wierz[0], wierz2[0], wierz1[0]) == 0 #zwracamy wtedy true gdy 3 wierzcholki tworza kat mniejszy od 180
    else:
        return angle_n_height(wierz1[0], wierz2[0], aktualny_wierz[0]) == 0

def triangulation(polygon):
    liczba_wierzcholkow = len(polygon)
    czesc_lewa, czesc_prawa = l_r_l(polygon) #podzial na 2 lancuchy
    uporzadkowana_lista = porzadkuj(czesc_lewa, czesc_prawa, liczba_wierzcholkow) #laczenie w 1 liste posortownana dla y-kow
    
    stos = []
    stos.append(uporzadkowana_lista[0])
    stos.append(uporzadkowana_lista[1])
    
    krawedzie = []
    for indeks in range(2, liczba_wierzcholkow): #ponizej sprawdzenie czy to ten sam lancuch
        if uporzadkowana_lista[indeks][2] != stos[-1][2]: #Jeśli kolejny wierzchołek należy do innego łańcucha niż wierzchołek stanowiący szczyt stosu, to możemy go połączyć ze wszystkimi wierzchołkamina stosie.
            for w in range(len(stos) - 1, -1, -1): #od konca stosu do 0 przechodzimy
                if abs(uporzadkowana_lista[indeks][1] - stos[w][1]) not in (1, liczba_wierzcholkow - 1): #unikamy dodawania krawędzi między wierzchołkami, które są sąsiadami w cyklicznym porządku (2 obok siebie albo ostatniego i pierwszego) sprawdzajac czy odleglosc indexow w polygon nie jest ani 1 ani n-1
                    krawedzie.append((uporzadkowana_lista[indeks][1], stos[w][1])) #laczymy wierzcholek z kazdym innym na stosie jesli nalezy do innego lancucha niz szczyt
            stos = [stos[len(stos)-1]]
        else:#Jeśli kolejny wierzchołek należy do tego samego łańcucha co wierzchołek ze szczytu stosu, to analizujemy kolejne trójkąty, jakie tworzy
#dany wierzchołek z wierzchołkami zdejmowanymi ze stosu to analizujemy kolejne trójkąty, jakie tworzy dany wierzchołek z wierzchołkami zdejmowanymi ze stosu:
            ostatni = len(stos) - 1
            for w in range(len(stos) - 2, -1, -1): #Jeśli trójkąt należy do wielokąta, to dodajemy przekątną i usuwamy odpowiedni wierzchołek ze szczytu stosu
                if trojkatdobry(stos[w], stos[ostatni], uporzadkowana_lista[indeks], liczba_wierzcholkow): 
                    krawedzie.append((uporzadkowana_lista[indeks][1], stos[w][1])) #Jeśli trójkąt należy do wielokąta, to dodajemy przekątną i usuwamy odpowiedni wierzchołek ze szczytu stosu
                    for k in range(len(stos) - 1, ostatni - 1, -1):
                        stos.pop()
                    ostatni = w
        stos.append(uporzadkowana_lista[indeks])
    return krawedzie #w przeciwnym przypadku umieszczamy zbadane wierzchołki na stosie albo raczej zbadane wierzchołki pozostają na stosie
````

````python
def triangulation_draw(polygon):
    liczba_wierzcholkow = len(polygon)
    czesc_lewa, czesc_prawa = l_r_l(polygon)
    uporzadkowana_lista = porzadkuj(czesc_lewa, czesc_prawa, liczba_wierzcholkow)
    
    stos = []
    stos.append(uporzadkowana_lista[0])
    stos.append(uporzadkowana_lista[1])
    
    krawedzie = []
    vis = Visualizer()
    for punkt in uporzadkowana_lista:
        vis.add_point(punkt[0], color="black")
        
    for i in range(len(polygon)):
        vis.add_line_segment(
            [polygon[i], polygon[(i + 1) % liczba_wierzcholkow]], color="gray"
        )
    
    
    
    for indeks in range(2, liczba_wierzcholkow):
        if uporzadkowana_lista[indeks][2] != stos[-1][2]:
            for w in range(len(stos) - 1, -1, -1):
                if abs(uporzadkowana_lista[indeks][1] - stos[w][1]) not in (1, liczba_wierzcholkow - 1):
                    krawedzie.append((uporzadkowana_lista[indeks][1], stos[w][1]))
                    vis.add_line_segment(
                        [uporzadkowana_lista[indeks][0], stos[w][0]],
                        color="green"
                    )
            stos = [stos[len(stos) - 1]]
        else:
            ostatni = len(stos) - 1
            for w in range(len(stos) - 2, -1, -1):
                if trojkatdobry(stos[w], stos[ostatni], uporzadkowana_lista[indeks], liczba_wierzcholkow): 
                    krawedzie.append((uporzadkowana_lista[indeks][1], stos[w][1]))
                    vis.add_line_segment(
                        [uporzadkowana_lista[indeks][0], stos[w][0]],
                        color="green"
                    )
                    for k in range(len(stos) - 1, ostatni - 1, -1):
                        stos.pop()
                    ostatni = w
        stos.append(uporzadkowana_lista[indeks])
    
    for krawedz in krawedzie:
        vis.add_line_segment([polygon[krawedz[0]], polygon[krawedz[1]]], color="green")
    
    return krawedzie, vis
````
